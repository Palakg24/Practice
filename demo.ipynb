BAR PLOT

```python
import matplotlib.pyplot as plt
x=["python","c++","c","java"]
y=[85,60,70,82]
z=[20,30,40,50]
plt.xlabel("languages",fontsize=15)
plt.ylabel("poplularity",fontsize=15)
plt.title("palak")
color=['r','y','b','m']
plt.bar(x,y,width=0.3,color='r',align="edge",edgecolor='black',alpha=0.8,label="palak")
plt.bar(x,z,width=0.3,color='y',align="edge",edgecolor='black',alpha=0.8,label="palak1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["python","c++","c","java"]
y=[85,60,70,82]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
plt.xlabel("languages",fontsize=15)
plt.ylabel("poplularity",fontsize=15)
plt.title("palak")
color=['r','y','b','m']
plt.bar(p,y,width,color='r',align="edge",edgecolor='black',alpha=0.8,label="palak")
plt.bar(p1,z,width,color='y',align="edge",edgecolor='black',alpha=0.8,label="palak1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","c++","c","java"]
y=[85,60,70,82]
z=[20,30,40,50]
plt.xlabel("languages",fontsize=15)
plt.ylabel("poplularity",fontsize=15)
plt.title("palak")
plt.barh(x,y,color='r',align="edge",edgecolor='black',alpha=0.8,label="palak")
plt.barh(x,z,color='y',align="edge",edgecolor='black',alpha=0.8,label="palak1")
plt.legend()
plt.show()
```

SCATTER PLOT

```python
import matplotlib.pyplot as plt
import numpy as np
day=np.array([1,2,3,4,5,6,7])
number=np.array([2,3,1,4,5,3,6])
no2=[3,4,7,6,9,1,2]
color=[10,49,30,29,56,20,30]
s=[100,200,700,900,110,300,400]
plt.scatter(day,number,c=color,s=s,cmap="autumn")
plt.scatter(day,no2,c='r',s=s)
t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.show()
```

PIE CHART

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([20,25,25,15])
y=np.array(["c","c++","java","python"])
ex=np.array([0.2,0.0,0.0,0.0])
c=['r','b','g','y']
plt.pie(x,labels=y,explode=ex,autopct="%0.0f%%",shadow=True,radius=0.7,labeldistance=1.1,startangle=0,textprops={"fontsize":11},
        wedgeprops={"linewidth":2,"edgecolor":'black'},rotatelabels=False)
plt.title("palak")
plt.legend(loc=4)
plt.show()
```
